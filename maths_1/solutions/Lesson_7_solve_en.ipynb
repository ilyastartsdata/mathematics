__6.__ Investigate for an extremum of a function:

### $$U=x^2y+\frac{1}{3}y^3+2x^2+3y^2-1$$

Solution:

Find the partial derivatives, equate them to zero and find the stationary points:

### $$\begin{cases}
   U'_{x}=2xy+4x=0 \\
   U'_{y}=x^2+y^2+6y=0 
 \end{cases}\,\,\,\,\,\Rightarrow\,\,\,\,\,
 \begin{cases}
   \left[ 
  \begin{gathered} 
     x=0\\
     y=-2
  \end{gathered} 
\right. \\
   x^2+y^2+6y=0 
 \end{cases}\,\,\,\,\,\Rightarrow$$
 
### $$\Rightarrow\,\,\,\,\,\left[ 
  \begin{gathered} 
     \begin{cases}
   x=0 \\
  y^2+6y=0 
 \end{cases}\\
     \begin{cases}
   y=-2 \\
   x^2=8 
 \end{cases}
  \end{gathered} 
\right.\,\,\,\,\,\Rightarrow\,\,\,\,\,
\begin{gathered} 
(0,0)\\
(0,-6)\\
(2\sqrt2,-2)\\
(-2\sqrt2,-2)
\end{gathered}$$
 
Find the second derivatives and form a Goesse matrix in general form:
 
### $$U''_{xx}=2y+4$$ 
### $$U''_{xy}=U''_{yx}=2x$$  
### $$U''_{yy}=2y+6$$

### $$\begin{pmatrix}
2y+4 & 2x\\ 
2x & 2y+6\\
\end{pmatrix}$$

you can now calculate the principal minors 

### $$\Delta_{1}=U''_{xx}=2y+4$$

### $$\Delta_{2}=\begin{vmatrix}
U''_{xx} & U''_{xy}\\ 
U''_{yx} & U''_{yy}\\
\end{vmatrix}=U''_{xx}\cdot U''_{yy}-(U''_{xy})^2=4y^2+20y+24-4x^2$$

It remains to check the minor signs for each point. 

For point $(0,0)$

### $$\Delta_{1}=2y+4=4>0$$

### $$\Delta_{2}=4y^2+20y+24-4x^2=24>0$$

then $(0,0)$ - minimum

For point $(0,-6)$

### $$\Delta_{1}=2y+4=-8<0$$

### $$\Delta_{2}=4y^2+20y+24-4x^2=48>0$$

then $(0,-6)$ - maximum

For point $(\pm2\sqrt2,-2)$

### $$\Delta_{1}=2y+4=0$$


As a rule, further investigation through the Goesse matrix makes no sense. But let's rebuild the Goesse matrix

### $$\begin{pmatrix}
U''_{yy} & U''_{yx}\\ 
U''_{xy} & U''_{xx}\\
\end{pmatrix}$$



### $$\begin{pmatrix}
2y+6 & 2x\\ 
2x & 2y+4\\
\end{pmatrix}$$

then 

### $$\Delta_{1}=U''_{yy}=2y+6=2\cdot(-2)+6=2>0$$

### $$\Delta_{2}=\begin{vmatrix}
U''_{yy} & U''_{yx}\\ 
U''_{xy} & U''_{xx}\\
\end{vmatrix}=U''_{xx}\cdot U''_{yy}-(U''_{xy})^2=4y^2+20y+24-4x^2=$$

### $$=4\cdot4+20\cdot(-2)+24-4\cdot8=40-40-32<0$$

then $(\pm2\sqrt2,-2)$ - saddles

```python
import warnings
warnings.filterwarnings('ignore')
import pylab
from mpl_toolkits.mplot3d import Axes3D
import numpy

def makeData ():
    x = numpy.arange (-8, 8, 0.1)
    y = numpy.arange (-8, 2, 0.1)
    xgrid, ygrid = numpy.meshgrid(x, y)

    Ugrid = xgrid**2*ygrid+1/3*ygrid**3+2*xgrid**2+3*ygrid**2-1
    return xgrid, ygrid, Ugrid

x, y, U = makeData()

fig = pylab.figure()

axes = Axes3D(fig)
# axes.view_init(azim=-30)
axes.plot_surface(x, y, U)

pylab.show()
```

__7*.__ Examine a function for an extremum:

### $$U=e^{-\frac{x}{2}}\cdot(x^2+y^2)$$

Solution:

Find the partial derivatives, equate them to zero and find the stationary points:

### $$\begin{cases}
   U'_{x}=-\frac{1}{2}e^{-\frac{x}{2}}(x^2+y^2)+e^{-\frac{x}{2}}\cdot2x=0 \\
   U'_{y}=e^{-\frac{x}{2}}\cdot2y=0 
 \end{cases}\,\,\,\,\,\Rightarrow$$
 
### $$\Rightarrow\,\,\,\,\,\begin{cases}
   -\frac{1}{2}(x^2+y^2)+2x=0 \\
   y=0 
 \end{cases}\,\,\,\,\,\Rightarrow\,\,\,\,\,
 \begin{cases}
   x^2-4x=0 \\
   y=0 
 \end{cases}\,\,\,\,\,\Rightarrow\,\,\,\,\,
 \begin{gathered} 
(0,0)\\
(4,0)
\end{gathered}$$
 
Find the second derivatives and form a Goesse matrix in general form:
 
### $$U''_{xx}=\frac{1}{4}e^{-\frac{x}{2}}(x^2+y^2)-\frac{1}{2}e^{-\frac{x}{2}}\cdot2x-\frac{1}{2}e^{-\frac{x}{2}}\cdot2x+2e^{-\frac{x}{2}}$$ 
### $$U''_{xy}=U''_{yx}=-\frac{1}{2}e^{-\frac{x}{2}}\cdot2y$$  
### $$U''_{yy}=2e^{-\frac{x}{2}}$$

It remains to check the minor signs for each point.

For point $(0,0)$

### $$\Delta_{1}=2>0$$

### $$\Delta_{2}=
\begin{vmatrix}
2 & 0\\ 
0 & 2\\
\end{vmatrix}=4>0$$

then $(0,0)$ - minimum

For point $(4,0)$

### $$\Delta_{1}=-\frac{2}{e^2}<0$$

### $$\Delta_{2}=
\begin{vmatrix}
-\frac{2}{e^2} & 0\\ 
0 & \frac{2}{e^2}\\
\end{vmatrix}=-\frac{4}{e^2}<0$$

then $(4,0)$ - saddles

```python
import pylab
import math
from mpl_toolkits.mplot3d import Axes3D
import numpy

def makeData ():
    x = numpy.arange (-8, 8, 0.1)
    y = numpy.arange (-8, 8, 0.1)
    xgrid, ygrid = numpy.meshgrid(x, y)

    Ugrid = math.e**(-xgrid/2)*(xgrid**2+ygrid**2)
    return xgrid, ygrid, Ugrid

x, y, U = makeData()

fig = pylab.figure()

axes = Axes3D(fig)
# axes.view_init(azim=-30)
axes.plot_surface(x, y, U)

pylab.show()
```

### $$y=\frac{1}{1+e^{-(ax+b)}}$$

Решение

```python
import numpy as np
import matplotlib.pyplot as plt


# Задаем рандомное распределение точек
xi = np.random.uniform(-8, 8, 200)
yi = 1/(1+np.exp(-np.random.uniform(0.5, 1.5, 200)*xi-np.random.uniform(-1, 1, 200)))
plt.plot(xi,yi, marker="o", ls="")


from scipy.optimize import fsolve


def equations(p):
    A, B = p
    # Это заранее найденные производные от (y-f(x))**2 по переменным a и b
    return ((-2*xi*np.exp(-A*xi-B)*(yi-1/(1+np.exp(-A*xi-B)))/(1+np.exp(-A*xi-B))**2).sum(), 
            (-2*np.exp(-A*xi-B)*(yi-1/(1+np.exp(-A*xi-B)))/(1+np.exp(-A*xi-B))**2).sum())

# Численное решение нелинечной системы уравнений
A, B =  fsolve(equations, (1, 1))
print (A, B)

x = np.linspace(-8, 8, 200)
plt.plot(x, 1/(1+np.exp(-A*x-B)))


plt.show()

```
